```python
import requests
from contextlib import closing
import csv
import pandas as pd
import xarray as xr
```

# Read in Valley Fever Data

```python
url = 'https://raw.githubusercontent.com/valleyfever/valleyfevercasedata/master/coccidioidomycosis_m2000_2015_v0.1.csv'

with closing(requests.get(url, stream=True)) as r:
    f = (line.decode('utf-8') for line in r.iter_lines())
    reader = csv.reader(f, delimiter=',')
    
    for _ in range(9):
        header = next(reader)
    time = pd.date_range(start=header[3], end=header[-1], freq=pd.offsets.MonthBegin())
    
    ds_dict = {} 
    for line in reader:
        state = line[0]
        county = line[1]
        fips = line[2]
        cases = [int(i) for i in line[3:]]
        
        combined_name = ( county + ', ' + state)
        
        da = xr.DataArray(cases, coords = [time], dims = ['time'], name = combined_name)
        da.attrs['state'] = state
        da.attrs['county'] = county

        ds_dict[combined_name] = da.to_dict()
    ds = xr.Dataset.from_dict(ds_dict)
```
